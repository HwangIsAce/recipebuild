```python
import pandas as pd

# sample = pd.read_csv('/home/donghee/projects/mlm/tester/test_output/10-next_token_predict_2023_1030_sample.tsv', sep='\t')
sample = pd.read_csv('/home/donghee/projects/mlm/tester/test_output/10-next_token_predict_2023_1030_all.tsv', sep='\t')
```

```python
sample.head()
```

```python
for i, v in enumerate(zip(sample['ANSWER_LIST'], sample['PRED_LIST_TEMP'])):
    if len(v[0].split(',')) != len(v[1].split(',')):
        print(i)
```

```python
_max = -1; answer_length = []
for i, v in enumerate(sample['ANSWER_LIST']):
    _length = len(v.split(','))
    answer_length.append(_length)
    if _length > _max:
        _max = _length
print(_max)
```

```python
import numpy as np

result = []
for i, v in enumerate(sample['ANSWER_LIST']):
    temp = []
    for j in range(_max):
        try:
            if eval(v)[j] in sample['PRED_LIST_TEMP'][i] :
                temp.append(1)
            else:
                temp.append(0)
        except:
            temp.append(np.nan)
    result.append(temp)
```

```python
score = pd.DataFrame(result)
col_name= []
for k in range(_max):
    col_name.append(f"pred_{k}")
score.columns = col_name

score['length'] = answer_length

recall_li = []
for l in range(len(score)):
    a = int(score.iloc[l][:_max].sum())
    b = int(score.iloc[l][-1])
    recall_li.append(a/b)
score['recall'] = recall_li

mean_li = list(score.mean())
score.loc[len(score)] = mean_li

score.tail()
```

```python
score.head()
```

```python
import statistics
macro_avg = statistics.mean(mean_li[:_max])
macro_avg
```

```python
_sum = 0
for i, v in enumerate(score.iloc[:-1, :_max].iterrows()):
    _sum += v[1].sum()

# micro_avg == recall 's avg ???
micro_avg = _sum / sum(answer_length)
micro_avg
```

```python
import pandas as pd

# type2 = pd.read_csv('/home/donghee/projects/mlm/tester/test_output/10-next_token_predict_2023_1027_sample2.tsv', sep='\t')
type2 = pd.read_csv('/home/donghee/projects/mlm/tester/test_output/10-next_token_predict_2023_1027_all2.tsv', sep='\t')
```

```python
type2.head()
```

```python
type2['PRED'] = type2['PRED'].apply(lambda x : eval(x))
```

```python
def make_recall(num):
    recall_n = []
    for i, v in enumerate(zip(type2['ANSWER'], type2['PRED'])):

        
        if v[0] in v[1][0:num]:
            recall_n.append(1)
        else :
            recall_n.append(0)

    return recall_n
    
```

```python
recall_1 = make_recall(1)
recall_3 = make_recall(3)
recall_5 = make_recall(5)
recall_10 = make_recall(10)
```

```python
score2 = pd.DataFrame({'ID': type2['IDX'], 'SUB_ID' : type2['SUB_IDX'], 'recall_1': recall_1, 'recall_3': recall_3, 'recall_5': recall_5, 'recall_10' :recall_10})
score2
```

```python
score2.mean() # macro avg
```

```python
micro_list = []
for i in range(len(score2['SUB_ID'].unique())): 
    i = i+1
    micro_list.append(list(score2[score2['SUB_ID'] == i].mean()[2:])) 
```

```python
micro_list_df = pd.DataFrame(micro_list)  
micro_list_df.columns = ['recall_1', 'recall_3', 'recall_5', 'recall_10']
print(micro_list_df.head())
print('')
print(len(micro_list_df))
```
